```python

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# --- 1. Create a 3x6 DataFrame with random float data ---
# This simulates having three data series (rows) over six time points or conditions (columns).
def drawInfluence(klist,data,dataname):
    columns = [f'k={i}' for i in klist]
    index = ['IMM','RTE','PRTE']
    df = pd.DataFrame(data, index=index, columns=columns)
    print("Generated DataFrame:")
    print(df)

    sns.set_style("whitegrid") # "whitegrid" is a clean style suitable for papers
    plt.figure(figsize=(10, 6), dpi=300)
    palette = sns.color_palette("viridis", n_colors=3)

    for i, series_name in enumerate(df.index):
        plt.plot(df.columns, df.loc[series_name],
                marker='o',           # Adds circles to each data point for clarity
                linestyle='-',        # Solid line style
                color=palette[i],     # Assign a color from the palette
                label=series_name)    # Label for the legend

    plt.title(f'Variation of influence with k on {dataname}', fontsize=16, fontweight='bold')
    plt.xlabel('Number of seeds', fontsize=12)
    plt.ylabel('Influence', fontsize=12)
    plt.xticks(fontsize=10)
    plt.yticks(fontsize=10)
    plt.legend(title='Method', fontsize=10, loc='best', frameon=True)
    plt.grid(True, which='both', linestyle='--', linewidth=0.5)
    plt.tight_layout()
    #plt.savefig('publication_quality_plot.png', dpi=300, bbox_inches='tight')
    plt.show()

def drawTime(klist,data,dataname):
    columns = [f'k={i}' for i in klist]
    index = ['IMM','RTE','PRTE']
    df = pd.DataFrame(data, index=index, columns=columns)
    print("Generated DataFrame:")
    print(df)

    sns.set_style("whitegrid") # "whitegrid" is a clean style suitable for papers
    plt.figure(figsize=(10, 6), dpi=300)
    palette = sns.color_palette("viridis", n_colors=3)

    for i, series_name in enumerate(df.index):
        plt.plot(df.columns, df.loc[series_name],
                marker='o',           # Adds circles to each data point for clarity
                linestyle='-',        # Solid line style
                color=palette[i],     # Assign a color from the palette
                label=series_name)    # Label for the legend

    plt.title(f'Variation of time with k on {dataname}', fontsize=16, fontweight='bold')
    plt.xlabel('Number of seeds', fontsize=12)
    plt.ylabel('Time (seconds)', fontsize=12)
    plt.xticks(fontsize=10)
    plt.yticks(fontsize=10)
    plt.legend(title='Method', fontsize=10, loc='best', frameon=True)
    plt.grid(True, which='both', linestyle='--', linewidth=0.5)
    plt.tight_layout()
    #plt.savefig('publication_quality_plot.png', dpi=300, bbox_inches='tight')
    plt.show()
```

```python
dataname = 'Soc-Epinions'
influenceList1=[654,706,859,1043,1198,1330]
influenceList2=[568,591,682,847,955,1072]
influenceList3=[637,722,836,1052,1201,1317]
klist=[20,30,50,100,150,200]
data = [influenceList1, influenceList2, influenceList3]
drawInfluence(klist,data,dataname)

timeList1=[106,139,228,268,387,552]
timeList2=[178,433,515,942,1281,1566]
timeList3=[67,68,68.9,69,69.4,70]
timeData = [timeList1, timeList2, timeList3]
drawTime(klist,timeData,dataname)
```
